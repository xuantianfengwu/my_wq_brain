# Combine Alphas

```python
import pandas as pd
factor_files = [
    # '../backfill/IND/[matrix]field_df_IND_TOP500_delay1.csv',
    # '../backfill/IND/[vector]field_df_IND_TOP500_delay1.csv',
    '../backfill/AMR/[field_df]AMR_TOP600_matrix_delay1.csv',
    '../backfill/AMR/[field_df]AMR_TOP600_vector_delay1.csv'
]
factor_df = pd.concat([ pd.read_csv(f) for f in factor_files ])
print(factor_df.shape)
factor_df.head()
```

```python
def extract_last_parentheses(text):
    """
    提取字符串中最后一个括号内的内容
    参数:
        text: 包含括号的字符串
    返回:
        最后一个括号内的内容，如果找不到括号返回None
    """
    # 查找最后一个 '(' 和最后一个 ')'
    left_idx = text.rfind('(')
    right_idx = text.find(')')
    # 确保找到两个括号且左括号在右括号之前
    if left_idx != -1 and right_idx != -1 and left_idx < right_idx:
        alpha = text[left_idx+1: right_idx].strip()
        alpha = alpha.split(',')[0].strip()
        alpha = alpha.strip('-')
        return alpha
    return None

file_ret_dict = {
    'rank': [
                '../backfill/AMR/analyst_matrix_D1_results.csv',

                '../backfill/AMR/shortinterest_vector_D1_results.csv',
                '../backfill/AMR/analyst_vector_D1_results.csv',
                '../backfill/AMR/earnings_vector_D1_results.csv',
                '../backfill/AMR/fundamental_vector_D1_results.csv',
                '../backfill/AMR/insiders_vector_D1_results.csv',
                '../backfill/AMR/institutions_vector_D1_results.csv',
                '../backfill/AMR/macro_vector_D1_results.csv',
                '../backfill/AMR/model_vector_D1_results.csv',
                '../backfill/AMR/news_vector_D1_results.csv',
                '../backfill/AMR/option_vector_D1_results.csv',
                '../backfill/AMR/other_vector_D1_results.csv',
                '../backfill/AMR/pv_vector_D1_results.csv',
                '../backfill/AMR/risk_vector_D1_results.csv',
                '../backfill/AMR/sentiment_vector_D1_results.csv',
             ],
    'delta5_rank': [],
    'delta60_rank': [],
    'backfill60_rank': [],
    'negetive_candidate': []
}
```

```python
final_values_df = pd.DataFrame()

for type in file_ret_dict:
    files = file_ret_dict[type]
    if len(files)>0:
        values_df = pd.concat([
            pd.read_csv(f).drop_duplicates().assign(file_name=f) for f in files
        ])
        print(type, values_df.shape)
        values_df['alpha'] = values_df['expression'].apply(extract_last_parentheses)
        final_values_df = pd.concat([final_values_df, values_df])

final_values_df = pd.merge(final_values_df,factor_df[['id','category_id']], how='left', left_on='alpha', right_on='id')
print('final_values_df', final_values_df.shape)

# Save final_values_df to CSV file
# final_values_df.to_csv('../backfill/IND/[backfill_result]D1_results.csv')
```

# generate_merge_df

```python
for category in final_values_df['category_id'].unique():
    # if category in ('pv'):
    if category in ('other'):
        cat_final_values_df = final_values_df[final_values_df['category_id'] == category]
        q_cat_final_values_df = cat_final_values_df[
                                        (cat_final_values_df['long_count']>=100) &
                                        (cat_final_values_df['sharpe']>1.2)
                                    ]
        print(category, q_cat_final_values_df.shape)

        import itertools
        expressions = q_cat_final_values_df['expression'].tolist()
        combinations = list(itertools.combinations(expressions, 2))
        print(f"总共生成 {len(combinations)} 个组合")

        # 迭代所有组合
        combine_expressions1 = [f'a={expr1};b={expr2};a+b'                                for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions2_1 = [f'a={expr1};b={expr2};ts_regression(a,b,20,rettype=1)'  for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions2_2 = [f'a={expr2};b={expr1};ts_regression(a,b,20,rettype=1)'  for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions3_1 = [f'a={expr1};b={expr2};ts_regression(a,b,20,rettype=2)'  for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions3_2 = [f'a={expr2};b={expr1};ts_regression(a,b,20,rettype=2)'  for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions4_1 = [f'trade_when({expr1}>0.5,{expr2},{expr1}<=0.5)'         for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions4_2 = [f'trade_when({expr2}>0.5,{expr1},{expr2}<=0.5)'         for i, (expr1, expr2) in enumerate(combinations)]
        combine_expressions = combine_expressions1 + combine_expressions2_1 + combine_expressions3_1 + combine_expressions4_1 + combine_expressions2_2 + combine_expressions3_2 + combine_expressions4_2
        alpha_list = [
            ace.generate_alpha(
                x,
                alpha_type="REGULAR",
                region="IND",
                universe="TOP500",
                delay=1,
                decay=5,
                test_period = "P2Y",
                neutralization = "INDUSTRY",
                truncation = 0.05,
                pasteurization = "ON",
            ) for x in combine_expressions
        ]
print(len(alpha_list))
```
